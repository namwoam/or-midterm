```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import os

mpl.rcParams['figure.dpi'] = 3000


def sort_df(df: pd.DataFrame) -> pd.DataFrame:
    df = df.sort_values("Data name")
    df["len"] = df["Data name"].str.len()
    df = df.sort_values("len", kind="mergesort")
    df = df.reset_index()
    return df


optimal_solution = pd.read_csv("optimal_result.csv")

optimal_solution = sort_df(optimal_solution)

huristic_solution = pd.read_csv("jen_result.csv")

huristic_solution = sort_df(huristic_solution)

print(optimal_solution.head())
print(huristic_solution.head())
```

```python
plt.plot(optimal_solution['Profit'], label='Optimal')
plt.plot(huristic_solution['Profit'], label='Huristic')

# add labels and title
plt.xlabel('Index')
plt.ylabel('Profit')
plt.title('Profit Comparison')

# add legend
plt.legend()

# show the plot
plt.show()


```
